#Reading the Dataset

```python
df = pd.read_csv("water_potability.csv")
df.head()
```

#Import library for Naive Bayes

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # for data visualization purposes
import seaborn as sns # for statistical data visualization
```

#Gaussian Model

#Informations of the dataset

```python
df.info()
```

#Drop NaN values

```python
df = df.dropna().reset_index(drop=True)
df
```

#Declare feature vector and target variable

```python
X = df.drop(['Potability'], axis=1)
y = df['Potability']
```

#Split data into separate training and test set 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# check the shape of X_train and X_test

X_train.shape, X_test.shape
```

#Feature Scaling 

```python
cols = X_train.columns
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])
```

```python
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
X_train.head()
```

#Model Training

```python
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB


# instantiate the model
gnb = GaussianNB()


# fit the model
gnb.fit(X_train, y_train)
```

#Predict the results 

```python
y_pred = gnb.predict(X_test)

y_pred
```

#Accuracy of the model

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.2f}'. format(accuracy_score(y_test, y_pred) *100))
```

Compare the train-set and test-set accuracy.

Now, I will compare the train-set and test-set accuracy to check for overfitting.

```python
y_pred_train = gnb.predict(X_train)

y_pred_train
```

```python
print('Training-set accuracy score: {0:0.2f}'. format(accuracy_score(y_train, y_pred_train) *100))
```

Check for overfitting and underfitting

```python
print('Training set score: {:.2f}'.format(gnb.score(X_train, y_train) *100))

print('Test set score: {:.2f}'.format(gnb.score(X_test, y_test)*100))
```

The values are really close, so there is no sign of overfitting

Overfitting: In train the model has good result but with test, he has a bad result

Underfitting: The model is bad in the trainig already, so there is no need of the test ocurrs.
