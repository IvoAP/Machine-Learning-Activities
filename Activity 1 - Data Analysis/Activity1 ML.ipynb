#2 Análise de Dados

Definição

Análise de Dados é um trabalho dedicado ao estudo e aprofundamento sobre um conjunto de dados coletados. Pois com a mesma somos capazes de descobrir padrões e tendências que podem fornecer informações valiosas sobre os dados.
Antigamente as análises de dados eram feitas em planilhas  ou por meio de relatórios o que gerou a necessidade de que os desenvolvedores precisassem otimizar a análise de dados por meio de gráficos mais intuitivos e algoritmos que facilitasse a obtenção da compreensão dos dados.

Tipos de Análises de dados

Análise Descritiva

É uma das análises mais comuns e é focada em usar os dados para se tomar decisões no presente e não para o futuro.
Ela utiliza da mineração de dados em tempo real visando encontrar respostas rápidas e seguras para as diversas questões do dia a dia das empresas.

Análise Preditiva

Utiliza fatos do passado para visualizar e prever eventos futuros. Ajudando as empresas a se protegerem de riscos e que as mesmas tenham um maior proveito de oportunidades. 
Para que esta seja feita de uma forma efetiva é necessário que sejam coletados os dados das mais diversas fontes para garantir um melhor cruzamento de informações.

Análise Prescritiva

Ela se assemelha um pouco a preditiva, porém diferente da preditiva que tenta prever o futuro, a prescritiva é focada em tentar determinar as consequências das decisões tomadas. Ou seja, ela se baseia em identificar as melhores estratégias, de acordo com os padrões existentes.

Análise Diagnóstica

Esta tem como objetivo realizar uma verificação mais ampla e geral sobre determinada situação. Como por exemplo, ela pode focar em traçar um perfil de comportamento do consumidor de uma determinada empresa para garantir melhores ações em relação ao marketing daquela empresa.

Etapas da Análise de Dados

Análise Exploratória

Podemos dizer que este primeiro  passo se baseia nas seguintes ações: Manusear se existem dados incompletos, depois verificarmos os pontos que estão fora da curva e por último iremos inserir os dados no sistema.

Modelagem de Dados

Neste segundo passo faz se necessário criar regras para os diferentes tipos de dados a se realizar e depois analisar a possibilidade de recursos de automação para facilitar a análise.

Construção de Relatórios

E o último passo será construir um relatório com base no que foi feito para que sejam tomadas as melhores decisões estratégicas com base na análise feita.
